```python
import pandas as pd
import time
chunksize = 1000000
columns_remain = ['overall', 'asin', 'style', 'reviewText', 'summary', 'vote']
electronic = pd.DataFrame()
start = time.time()
for chunk in pd.read_json("Electronics.json", lines=True, chunksize=chunksize):
    electronic = pd.concat([electronic, chunk[columns_remain]], ignore_index=True)
end = time.time()
print("Execute time: {}".format(end - start))
```

```python
electronic
```

```python
electronic = electronic.dropna()
```

```python
electronic = electronic.drop(['style'], axis=1)
```

```python
electronic
```

## summary + reviewText->text

```python
def get_text(row):
    return "{} {}".format(row['summary'], row['reviewText'])

electronic['text'] = electronic.apply(get_text, axis=1)
electronic
```

```python
electronic = electronic.drop(['reviewText', 'summary'], axis=1)
electronic
```

## 将asin映射为ID

```python
electronic['asin_id'] = pd.factorize(electronic['asin'])[0]
asin_to_id = electronic[['asin', 'asin_id']].drop_duplicates().set_index('asin')
asin_to_id
```

```python
electronic = electronic.drop(['asin'], axis=1)
electronic
```

```python
import re
def clean(text):
    # 替换HTML标签和空白
    text = re.sub('\s+', ' ', text)  # 去除多余空白
    return text
electronic['text'] = electronic['text'].apply(lambda x : clean(x))
electronic
```

```python
electronic.to_csv("electronic_review_cleaned.csv", index=None)
```
