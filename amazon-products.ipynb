```python
import numpy as np
import pandas as pd
```

```python
chunksize = 500000
children = pd.DataFrame()
for chunk in pd.read_json("meta_Books.json", lines=True, chunksize=chunksize):
    children = pd.concat([children, chunk], ignore_index=True)
# children = pd.read_json("meta_Books.json", lines=True, chunksize=chunksize)
```

```python
children.tail()
```

```python
columns_remain = ['category', 'description', 'title', 'also_buy', 'also_view', 'price', 'asin']
remain = children[columns_remain]
```

```python
remain['class'] = remain['category'].apply(lambda x : False if len(x) <= 1 else True)
remain = remain.drop(remain[remain['class'] == False].index)
remain['class'] = remain['category'].apply(lambda x : False if x[1] != 'History' else True) #TBA
remain = remain.drop(remain[remain['class'] == False].index)
remain['class'] = remain['category'].apply(lambda x : False if len(x) <= 2 else True)
remain = remain.drop(remain[remain['class'] == False].index)
```

```python
remain['category'] = remain['category'].apply(lambda x : x[2])
```

```python
remain
```

```python
remain = remain.reset_index(drop=True)
# remain['class'] = remain['description'].apply(lambda x : False if len(x) == 0 else True)
# remain = remain.drop(remain[remain['class'] == False].index)
remain['class'] = remain['description'].apply(lambda x: True if len([i for i in x if i]) > 0 else False)
remain = remain.drop(remain[remain['class'] == False].index)

remain['class'] = remain['title'].apply(lambda x: True if len([i for i in x if i]) > 0 else False)
remain = remain.drop(remain[remain['class'] == False].index)
```

```python
remain = remain.reset_index(drop=True)
remain
```

```python
def extract_text(lst):
    lst = [s for s in lst if s != '']
    return ''.join(lst)


remain['cleaned_desc'] = remain['description'].apply(extract_text)
remain
```

```python
def get_value(row):
    return "Description: {0}; Title: {1}".format(row['cleaned_desc'], row['title'])

remain['text'] = remain.apply(get_value, axis=1)
remain = remain.drop(['description', 'title', 'class'], axis=1)
remain
```

```python
import re
def clean(text):
    # 替换HTML标签和空白
    text = re.sub('<[^<]+?>', '', text)  # 去除HTML标签
    text = re.sub('\s+', ' ', text)  # 去除多余空白
    return text
remain['text'] = remain['text'].apply(lambda x : clean(x))
remain
```

```python
remain = remain.reset_index(drop=True)
# remain.reset_index(drop=True)
out = remain[['category','also_buy','also_view', 'price', 'asin', 'text']]
```

## Feature Engineering

```python
remain = pd.read_csv("cleaned.csv")
```

```python
remain.reset_index()
```

```python
remain
```

```python
categories = list(remain['category'].unique())
old_dict = dict(enumerate(categories))
class_to_number = {value: key for key, value in old_dict.items()}

remain['label'] = remain['category'].map(class_to_number)
# remain
```

```python
remain['asin_id'] = pd.factorize(remain['asin'])[0]
asin_to_id = remain[['asin', 'asin_id']].drop_duplicates().set_index('asin') #去重过的
df_filtered = remain[remain['asin'].isin(asin_to_id.index)]
df_filtered.drop_duplicates(subset = ['asin'], keep='last', inplace=True)
```

```python
asin_to_id
```

```python
import ast
my_list = ast.literal_eval(df_filtered['also_buy'][2])
```

```python
df_filtered['also_buy'] = df_filtered['also_buy'].apply(lambda x : ast.literal_eval(x))
```

```python
df_filtered['also_view'] = df_filtered['also_view'].apply(lambda x : ast.literal_eval(x))
```

```python
df_filtered['also_buy'] = df_filtered['also_buy'].apply(lambda x: [asin for asin in x if asin in asin_to_id.index])
df_filtered['also_buy'] = df_filtered['also_buy'].apply(lambda x: [asin_to_id.loc[asin, 'asin_id'] for asin in x])
df_filtered['also_view'] = df_filtered['also_view'].apply(lambda x: [asin for asin in x if asin in asin_to_id.index])
df_filtered['also_view'] = df_filtered['also_view'].apply(lambda x: [asin_to_id.loc[asin, 'asin_id'] for asin in x])
```

```python
df_filtered
```

```python
df_filtered = df_filtered.reset_index()
```

```python
new_df = df_filtered
new_df['node_id'] = pd.factorize(new_df['asin_id'])[0]
new_df['merge'] = new_df['also_view'] + new_df['also_buy']
new_df['merge'] = new_df['merge'].apply(lambda x : list(set(x)))
new_df['merge'] = new_df['merge'].apply(lambda x : sorted(x))
```

```python
adj_list = {}
asin_ids = sorted(new_df['node_id'].unique())
for asin_id in asin_ids:
    adj_list[asin_id] = []
for index, row in new_df.iterrows():
    src_asin_id = row['node_id']
    merge_asins = row['merge']
    for dest_asin in merge_asins:
        adj_list[src_asin_id].append(dest_asin)
```

```python
pairs = []
adj0 = []
adj1 = []
for src_asin_id, neighbors in adj_list.items():
    for dest_asin_id in neighbors:
        adj0.append(src_asin_id)
        adj1.append(dest_asin_id)
```
